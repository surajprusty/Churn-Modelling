```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13]
y = dataset.iloc[:, 13]
```

```python
#Create dummy variables
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
```

```python
## Concatenate the Data Frames
X=pd.concat([X,geography,gender],axis=1)
```

```python
## Drop Unnecessary columns
X=X.drop(['Geography','Gender'],axis=1)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Part 2 - Now let's make the ANN!
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential #Sequential library is used for Neural Networks(ANN,CNN,RNN)
from keras.layers import Dense # Dense is used for creating hidden layers
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(units=6, kernel_initializer = 'he_uniform',activation='relu',input_dim = 11)) # units is for neurons in first hidden layer
#kernel_initializer is used for weight intialization and input_dim is used for no.of innput parameters
```

```python
# Adding the second hidden layer
classifier.add(Dense(units= 6, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])#optmizer is used for reducing the cost function and will for updating the weights(error between prediction and actual value)
#For binary classification we used binary_crossentropy whereas for non-binary classification or than two we will use "categorical_crossentropy"
#For non binary classification, the activation in output will not be 'sigmoid',but 'softmax'
```

```python
# Fitting the ANN to the Training set
model_history=classifier.fit(X_train, y_train,validation_split=0.33, batch_size = 10, epochs = 100) #Batch size is used for improving computational power becomes less
```

```python
# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
